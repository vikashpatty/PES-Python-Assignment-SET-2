Use Bubble sort / Selection sort technique to sort the elements of List) 

Note: Don't use built in functions to sort.

```python
def bubbleSort(arr): 
    n = len(arr) 
    for i in range(n): 
        for j in range(0, n-i-1):  
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 

arr = [9,1,3,6,77,8]
bubbleSort(arr)
print(arr)
```
